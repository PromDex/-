```python
import numpy as np
from math import factorial as f
```

### подготовим необходимые формулы и оформим их как вызываемые функции

```python
# Формула сочетания
def combinations(n, m):
    return int(f(n) / (f(m) * f(n - m)))
```

$$C_n^m = \frac{n!}{m!(n - m)!}$$

```python
# Формула Бернулли
def bern(p, n, k):
    return combinations(n,k) * p**(k) * (1 - p)**(n - k)
```

$$p(k) = C_n^kp^kq^{n-k}$$

```python
# Формула Пуассона
def puas(m, n, p):
    λ = n * p 
    return (λ**m / f(m)) * np.exp(-λ)
```

$$P_n(k)=\frac{\lambda^k}{k!}\cdot e^{-\lambda}$$

**ЗАДАЧА 1.**\
Стрелок выстрелил 100 раз.\
Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8.\
Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

**решение**\
производено опытов :   n = 100\
вероятность наступления события A в каждом из которых:   p = 0.8\
найти вероятность, что событие A наступит в точности:   k = 85 раз

```python
n = 100
p = 0.8
k = 85

# вызов функции - формула Бернулли.
P = bern(p, n, k)
print(f'Ответ: вероятность того, что стрелок попадет в цель ровно 85 раз: {P :.3f}')
```
